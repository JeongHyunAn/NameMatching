```python
import sys
from csv import reader

def twoSurname(f_name, s_name, s_len):
    temp = ['t']
    f_name = f_name + temp
    f_name[-1] = s_name[s_len - 2]
    f_name = f_name + temp
    f_name[-1] = s_name[-1]
    return f_name

def oneSurname(f_name, s_name):
    temp = ['t']
    f_name = f_name + temp
    f_name[-1] = s_name[-1]
    return f_name


test_file = sys.argv[1] # Read from command line
count = 0
output = open('full-name-output.csv' , 'w')
test_lines = open('dev_test.csv' , 'r')
with open('name_not_included.csv', 'r') as f:
    n_name_list = f.read().splitlines()
    
for line in test_lines:
    flag = False
    line = line.strip() # Remove newline character
    [first_person , second_person] = line.split(' AND ')
    first_person = list(map(str, first_person.split()))
    second_person = list(map(str, second_person.split()))
    f_len = len(first_person)
    s_len = len(second_person)

    #1
    if f_len == 1 and s_len == 2:
        first_person = oneSurname(first_person, second_person)
    
    elif f_len == 1 and s_len == 3:
        first_person = oneSurname(first_person, second_person)
        
    elif f_len == 1 and s_len == 4:
        first_person = twoSurname(first_person, second_person, s_len)

    #2
    elif f_len == 2 and s_len == 1:
        first_person = oneSurname(first_person, second_person)
        
    elif f_len == 2 and s_len == 2:
        for i in range(len(n_name_list)):
            if first_person[-1] == n_name_list[i]:
                first_person = oneSurname(first_person, second_person)
        
    elif f_len == 2 and s_len == 3:
        for i in range(len(n_name_list)):
            if first_person[-1] == n_name_list[i]:
                first_person = twoSurname(first_person, second_person, s_len)
                
        first_person = oneSurname(first_person, second_person)
    
    elif f_len == 2 and s_len == 4:
        for i in range(len(n_name_list)):
            if first_person[-1] == n_name_list[i]:
                first_person = twoSurname(first_person, second_person, s_len)

    predicted_first_person = " ".join(first_person)
    output.write (line + ',' + predicted_first_person + '\n')
output.close

with open('dev_key.csv', 'r') as f:
    key_name = reader(f)

    key_list = [row for row in key_name]
    
with open('full-name-output.csv', 'r') as f:
    test_name = reader(f)

    test_list = [row for row in test_name]

for i in range(len(test_list)):
    if test_list[i][1] == key_list[i][1]:
        count += 1
print(count/len(test_list))
```
